```python
#searched for a review that stop me from finding the average reviews
df[df['Reviews']== '3.0M']
```

```python
#removed the 'm' from the data
df['Reviews']=df['Reviews'].replace('3.0M',3.0)
```

```python
#changed the data type to float
df['Reviews']=df['Reviews'].astype('float')
```

```python
#found the average number of the reviews
round(df['Reviews'].mean(),2)
```

```python
#how many apps are free,how many are paid?
df['Type'].value_counts()
```

```python
#Which app has the most reviews
df[df['Reviews'].max()==df['Reviews']]
```

```python
#what are the top 5 apps in reviews?
df.groupby("App")['Reviews'].max().sort_values(ascending=False).head()
```

```python
#what is the average rating by type of app
df.groupby('Type')['Rating'].mean()
```

```python
#looking at installs column
df['Installs']
```

```python
#creating new installs column without commas
df['Installs_1']=df['Installs'].str.replace(',','')
```

```python
df.head(1)
```

```python
#removing '+' from the new installs column
df['Installs_1']=df['Installs_1'].str.replace('+','')
```

```python
df.head(1)
```

```python
#attempt to convert new installs column to int
df['Installs_1'].astype('int')
```

```python
# finding the row with the word free
df[df['Installs_1']=='Free']
```

```python
#replacing the word 'free' with 0
df['Installs_1']=df['Installs_1'].str.replace('Free','0')
```

```python
#converting new installs column to int
df['Installs_1']=df['Installs_1'].astype('int')
```

```python
#creating a variable that is defined by the top 5 apps with the most installs
index=df['Installs_1'].sort_values(ascending=False).head().index
```

```python
#using the previous variable to display the names of the apps with the most installs
df.iloc[index]['App']
```
