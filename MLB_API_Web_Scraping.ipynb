```python
import pandas as pd
import requests
pd.set_option('display.max_columns',None)
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
test_url = 'https://bdfed.stitch.mlbinfra.com/bdfed/stats/team?&env=prod&sportId=1&gameType=R&group=hitting&order=desc&sortStat=homeRuns&stats=season&season=2024&limit=30&offset=0'
```

```python
df=requests.get(url= test_url).json()
df
```

```python
df=pd.DataFrame(df['stats'])
```

```python
df_backup=df.copy()
df_backup
```

```python
df.to_excel('MLB_team_data.xlsx',index=False)
```

```python
#dropping columns that are not necessary
df=df.drop(columns=['year','type','rank','shortName','teamId','teamAbbrev','teamShortName','leagueName','leagueShortName','gamesPlayed'])
df.columns.tolist()
```

```python
#which team had the most runs?
df[df['runs'].max()==df['runs']]
```

```python
#which teams were top 5 in triples?
df.groupby('teamName')['triples'].max().sort_values(ascending=False).head()
```

```python
#which teams were top 5 in doubles?
df.groupby('teamName')['doubles'].max().sort_values(ascending=False).head()
```

```python
#which teams were above average at stealing bases
df[df['stolenBases']>(df['stolenBases'].mean())][['teamName','stolenBases']]
```

```python
#importing libraries for graphing
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

plt.figure(figsize=(10,10))
plt.xlabel('Team')
plt.ylabel('Home Runs')
plt.bar(df['teamName'], df['homeRuns'], color=['green','green','green','green','black','green','green','black','black','green','green','green','black','black','black','green','black','black','black','black','green','black','black','green','black','black','black','black','black','black'])
plt.xticks(rotation=90)
plt.legend(['Made Playoffs','Missed Playoffs'],loc='upper right')
plt.show()
```

```python
#create variable for numerical columns
numerical_columns=df.select_dtypes(include=['int64','float']).corr()
```

```python
#using previous varible to create correlation chart
plt.figure(figsize=(15,10))
sns.heatmap(numerical_columns,annot=True)
plt.title('correlation')
plt.show()
```

```python
sns.catplot(data=df, x='strikeOuts',y='leagueAbbrev',kind='box',hue='leagueAbbrev', height=7)
```

```python
numerical_columns2 = df.select_dtypes(include=["int64", "float64"]).columns

# Plot distribution of each numerical feature
plt.figure(figsize=(14, len(numerical_columns2) * 3))
for idx, feature in enumerate(numerical_columns2, 1):
    plt.subplot(len(numerical_columns2), 2, idx)
    sns.histplot(df[feature], kde=True)
    plt.title(f"{feature} | Skewness: {round(df[feature].skew(), 2)}")

# Adjust layout and show plots
plt.tight_layout()
plt.show()
```
